```python
#importing libraries, I will use for cleaning and modifying the dataset
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#reading csv file. Applied set_option to see full data in the series 
df = pd.read_csv(r'C:\Users\v-vovnenko\Desktop\python\us-software-engineer-jobs.csv', sep = ';')
pd.set_option('display.max_colwidth', 3000)
df
```

```python
#checking an amount of rows snd columns in the dataframe
df.shape
```

```python
#droping the rows with NaN values in all series + a row without a value in a first serie
#resetting indexes after dropiing NaN rows
df.drop(df.index[36:47], inplace=True)
df.drop(df.index[3], inplace=True)
df.reset_index(drop = True, inplace=True)
df
```

```python
#replacing 'Sr.' to 'Senior' in the column 'title'
df['title'].replace('Sr.', 'Senior', regex = True)
```

```python
#choosing all titles which start from 'Junior', 'Middle', Senior'
Junior_Middle_Senior_titles = df['title'].str.contains('Junior|Middle|Senior')
titles = df[Junior_Middle_Senior_titles]
titles
```

```python
df['seniority level'] = titles['title'].str.split(' '). str[0]
```

```python
#df.dtypes
df['seniority level'].memory_usage()
```

```python
#memory dropped from 412 to 371
df['seniority level'] = df['seniority level'].astype('category')
df['seniority level'].memory_usage()
```

```python
df['company'] = df['company'].str.title()
df
```

```python
#replacing a symbol '-' with ''
df['rating'] = df['rating'].astype(str).str.replace('-', '')
df
```

```python
#removing state in the column 'location'
df['location'] = df['location'].str.split(',').apply(lambda x: x[0])
df
```

```python
df = df.drop(['Unnamed: 14'], axis = 1)
df.head()
```
